```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
basket_df = pd.read_csv("basket.csv")
basket_df.head()
```

```python
basket_df.tail()
```

```python
transactions = []
for i in range(0,14963):
    transactions.append([str(basket_df.values[i,j])for j in range(0,11)])
```

```python
from apyori import apriori
rules = apriori(transactions=transactions, min_support=0.0001, min_confidence=0.6, min_lift= 3, min_length =2, max_length = 3)
```

```python
results=list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame
```

```python
resultsinDataFrame=resultsinDataFrame.drop_duplicates()
```

```python
resultsinDataFrame.sort_values(by="Lift",ascending=False)
```

# MLXTEND İLE APRİORİ

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


transactions = []
for i in range(0,14962):
    transactions.append([str(basket_df.values[i,j])for j in range(0,11)])

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)


frequent_itemsets = apriori(df, min_support=0.0001, use_colnames=True)


rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```

```python
frequent_itemsets_df = pd.DataFrame(frequent_itemsets)
rules_df = pd.DataFrame(rules)
```

```python
frequent_itemsets_df=frequent_itemsets_df.drop_duplicates()
rules_df=rules_df.drop_duplicates()
```

```python
frequent_itemsets_df.sort_values(by="support")
```

```python
rules_df.sort_values(by="lift",ascending=False)
```
