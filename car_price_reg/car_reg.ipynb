```python
def check_dataset(dataset):
    print("İlk 5 satır")
    print(dataset.head())
    print("_" * len(dataset))
    
    print("Son 5 satır")
    print(dataset.tail())
    print("_" * len(dataset))
    
    print("Genel bilgi")
    dataset.info()  
    print("_" * len(dataset))
    
    print("Şekli")
    print(dataset.shape)
    print("_" * len(dataset))
    
    print("İstatistiksel değerler")
    print(dataset.describe().T)
    print("_" * len(dataset))
    
    print("Eksik değer sayısı")
    print(dataset.isnull().sum())
    print("-" * len(dataset))
    
    print("Kategorik ama kardinal")
    cat_but_car = [col for col in dataset.columns if dataset[col].nunique() > 20 and dataset[col].dtype == "object"]
    print(cat_but_car)
    print("-" * len(dataset))
    
    print("Sayısal ama kategorik değerler")
    num_but_cat = [col for col in dataset.columns if dataset[col].nunique() < 10 and dataset[col].dtype in ["int64", "float64"]]
    print(num_but_cat)
    print("-" * len(dataset))
    
    print("Kategorik değerler")
    cat_cols = [col for col in dataset.columns if dataset[col].dtype in ["object", "category", "bool"]]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    print(cat_cols)
    print("-" * len(dataset))
    
    print("Sayısal değerler")
    num_cols = [col for col in dataset.columns if dataset[col].dtype in ["int64", "float64"]]
    num_cols = [col for col in num_cols if col not in cat_cols]
    print(num_cols)
    print("-" * len(dataset))
    
    print("Eşsiz değerler")
    print(dataset.nunique())
    print("-" * len(dataset))
```

```python
check_dataset(df)
```
